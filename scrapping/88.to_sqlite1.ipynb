```python
import sqlite3
import pandas as pd
```

```python
db_news = sqlite3.connect('./news.db')
c = db_news.cursor()
```

```python
c.execute("CREATE TABLE newspapers (id INTEGER PRIMARY KEY, title TEXT, press TEXT, category TEXT, date TEXT, name TEXT, text TEXT)")
```

```python
df = pd.read_excel('./sample.xls', names=['title', 'press', 'category', 'date', 'name', 'text'])
df
```

```python
contents = df.values.tolist()
contents
```

```python
id = 1
for content in contents:
    title = content[0]
    press = content[1]
    category = content[2]
    date = content[3]
    name = content[4]
    text = content[5]
    c.execute("INSERT INTO newspapers (id, title, press, category, date, name, text) VALUES(?,?,?,?,?,?,?) ",(id, title, press, category, date, name, text))
    id +=1
db_news.commit()
```
