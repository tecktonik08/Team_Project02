```python
def search_news(text, num):
  searched_articles = []

  for i in range(num):
    if text not in titles[i]:
      continue
    else:
      searched_articles.append(news[i])
  
  for i in range(num):
    if text not in keywords[i]:
      continue
    else:
      searched_articles.append(news[i])

  return searched_articles
```

```python
from konlpy.tag import Mecab
mecab = Mecab()
```

```python
# 의미 없는 단어(조사 등) 제외
stopwords = ['은', '도', '는', '다', '사', '서', '의', '위', '제', '회', '등', '가', '이', '했', '한', '에', '애', '하', '보', '웃', '고', '곧', '정', '적', '순', '뜻', '용', '열', '유', '뜻', '잘', '못', '타', '희', '히', '점', '을', '를', '인', '듯', '과', '와', '네', '들', '듯', '지', '임', '게', '개', '대', '것', '만', '명', '원', '세','에서', '이고', '라고', '다고', '라기', '라며', '면서', '라면서', '로써', '로서', '로', '으로', '뿐', '에서', '어야', '부터', '었', '았', '한다', '이다', '였다', '였었다', '|', '/','‘', '’', ',' , '“', '”', '.', '>', '<', ')', '(']
```

```python
import re
```

```python
def mec(i):
  words = list()
  for tok in news[i]:
    temp = ''.join(tok)
    temp_re = re.sub('[^가-힣ㄱ-ㅎㅠ ]','', temp)
    encoded = mecab.morphs(temp_re)
    wd = [item for item in encoded if item not in stopwords]  # ('[^가-힣ㄱ-ㅎㅠ ]','')
    words.append(wd)

  return words
```
